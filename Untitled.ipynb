```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
t = np.linspace(0, 10, 51, endpoint = True)
t
```

```python
f = np.cos(t)
f
```

```python
np.cos(1)
```

```python
plt.plot(t,f, color="green")
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3, 3, 51, endpoint = True)
x
```

```python
y1 = x**2
plt.plot(x, y1)
```

```python
y2 = 2 * x + 0.5
plt.plot(x, y2)
```

```python
y3 = -3 * x - 1.5
plt.plot(x, y3)
```

```python
y4 = np.sin(x)
plt.plot(x, y4)
```

```python
fig = plt.figure()
fig.set_size_inches(8, 6)
ax = [0]*4
for i in range(1, 5):
    ax[i-1] = fig.add_subplot(2, 2, i)
    ax[i-1].set_title('График y' + str(i))
    title = 'y'+str(i)
    ax[i-1].plot(x, globals()[title])

ax[0].axis([-5, 5, min(y1), max(y1)])

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
for i in range(4):
    title = 'y'+str(i+1)
    ax[i].plot(x, title)
    plt.show()
```
